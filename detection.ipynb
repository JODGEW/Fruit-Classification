```python
import cv2
import torch
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('../apple_test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(16, 16))
plt.imshow(img)
plt.title('Fruits')
plt.xticks([])
plt.yticks([])
```

```python
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.xticks([])
plt.yticks([])
```

```python
bin_gray_img = cv2.threshold(gray_img, 245, 255, cv2.THRESH_BINARY)

plt.imshow(bin_gray_img[1], cmap='gray')
plt.title('Binary version of apple image')
plt.xticks([])
plt.yticks([])
```

```python
inv_bin_gray_img = ~bin_gray_img[1]

plt.imshow(gray_img, cmap='gray')
plt.title('Inverted Binary version of apple image')
plt.xticks([])
plt.yticks([])
```

```python
contours_list, heirachy = cv2.findContours(inv_bin_gray_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

with_contours = cv2.drawContours(img, contours_list, -1, (255, 0, 0), 5)

plt.imshow(img)
plt.title('Image with red boundary')
plt.xticks([])
plt.yticks([])
```

```python
print('Number of contours in image:', len(contours_list))
```

```python
contour = contours_list[0]
```

```python
many_img = cv2.imread('../apple.jpg')
many_img = cv2.cvtColor(many_img, cv2.COLOR_BGR2RGB)
fig = plt.figure(figsize=(16, 16))
plt.imshow(many_img)
plt.title('Fruits')
plt.xticks([])
plt.yticks([])
```

```python
many_gray_img = cv2.cvtColor(many_img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap='gray')
plt.xticks([])
plt.yticks([])
```

```python
many_bin_gray_img = cv2.threshold(many_gray_img, 245, 255, cv2.THRESH_BINARY)

plt.imshow(many_bin_gray_img[1], cmap='gray')
plt.title('Binary version of apple image')
plt.xticks([])
plt.yticks([])
```

```python
many_inv_bin_gray_img = ~many_bin_gray_img[1]

plt.imshow(many_inv_bin_gray_img, cmap='gray')
plt.title('Inverted Binary version of apple image')
plt.xticks([])
plt.yticks([])
```

```python
contours, heirachy = cv2.findContours(many_inv_bin_gray_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)

with_contours = cv2.drawContours(many_img, contours, -1, (255, 0, 0), 5)

plt.imshow(img)
plt.title('Image with red boundary')
plt.xticks([])
plt.yticks([])
```

Many fruits detection

```python
match_distance_list = []

for cnt in contours:
    match_distance = cv2.matchShapes(cnt, contour, 1, 0)
    match_distance_list.append(match_distance)
```

```python
sorted_list = match_distance_list.copy()
```

```python
sorted_list.sort()
```

```python
ind1_dist = match_distance_list.index(sorted_list[0])

ind2_dist = match_distance_list.index(sorted_list[1])
```

```python
apple_cnts = []
apple_cnts.append(contours[ind1_dist])
apple_cnts.append(contours[ind2_dist])
```

```python
with_contours = cv2.drawContours(many_img, apple_cnts, -1, (0, 0, 255), 3)

fig = plt.figure(figsize=(16, 16))
plt.imshow(many_img)
plt.title('Detection of apple with blue boundary')
plt.xticks([])
plt.yticks([])
```
